```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
```

```python
path_saved_model = 'model_mobilenet_acc81.h5'

loaded_model = load_model(path_saved_model)
loaded_model.summary()
```

```python
test_datagen = image.ImageDataGenerator(
    rescale = 1./255
)

test_data = test_datagen.flow_from_directory(
    'dataset/test',
    target_size = (224, 224),
    batch_size = 16,
    class_mode = 'categorical'
)
```

```python
loaded_model.evaluate(test_data)
```

```python
test_data.class_indices
```

```python
img_path = 'dataset/test/rendang/rendang-21979.jpg'

img = image.load_img(img_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)

classes = loaded_model.predict(img_array/255)

for i, class_ in enumerate(classes[0]):
    if class_ == classes[0].max():
        print(i)
```
